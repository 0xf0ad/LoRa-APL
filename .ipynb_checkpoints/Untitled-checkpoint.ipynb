```apl
]dinput
fft ← {
    M    ← ≢⍵
    k    ← 2÷⍨M
    1>K: ⍵
    0≠1 | 2⍟M: 'M nest pas un multiple de 2'
    odd  ← ∇(M⍴1 0)/⍵
    even ← ∇(M⍴0 1)/⍵
    exp  ← *(0J¯2×(○1)×(¯1+⍳K)÷M)
    T    ← even × exp
    | (odd+T), odd-T
}
```

```apl
]dinput
plot ← {
    do ← {⍎'⍺⍺ ⍵ ⋄ ⍵' ⋄ ⍺⍺}
    _  ← InitCauseway do ⍬
    sp ← ⎕NEW Causeway.SharpPlot
    sp.LineLimit ← 0
    _  ← sp.DrawLineGraph do ⍵
    svg ← sp.RenderSvg Causeway.SvgMode.FixedAspect
    _  ← 'Reaction Times' (3500⌶) svg
}
```

```apl
msg ← 'Hello, world' ⍝ le mot a moduler
SF  ← 8              ⍝ spread factor
ns  ← 2*16           ⍝ nombre de samples
B   ← 125e3          ⍝ bandwitdh
f0  ← 433e6
T   ← 1 ÷ B          ⍝ periode
Ts  ← T × 2*SF       ⍝ periode de symbole
k   ← ¯1 + ⍳2*SF     ⍝ variable du temps
t   ← (¯1 + ⍳ns) × Ts ÷ ns
SNR ← ¯0
Ps  ← 1
```

```apl
]dinput
chirp ← {
    Re ← { (⍵ + +⍵) ÷ 2 }
    s  ← ⍵
    f  ← ((2*SF) | (t÷T) + s) × B÷2*SF
    phi ← +\ f × Ts÷ns
    c  ← Re *(0J1 × ○2 × phi)
    ⍝ c (2×phi÷(B×Ts)) (f ÷ B)
    c
}
```

```apl
]dinput
chirp_string ← {
    w ← ⎕UCS ⍵
    ∊chirp¨w
}
```

```apl
]dinput
gauss ← {                                 ⍝ Random numbers with a normal distribution
   depth ← 2*32                           ⍝ profondeure de aletoirement
   (x y) ← ⊂[1+⍳⍴,⍵](?(2,⍵)⍴depth)÷depth  ⍝ two random variables within ]0;1]
   ((¯2×⍟x)*0.5)×1○○2×y                   ⍝ methode de Box-Muller
}
```

```apl
plot (chirp 231) t
c ← chirp 231
plot (c + (10*SNR÷¯10) × (Ps*0.5) × gauss ≢c) t
⍝ plot fft ⊃chirp 23
⍝ plot (chirp_string 'Hello')
⍝ plot ((chirp_string msg))
⍝ plot (f÷B)  (t÷T)
⍝ freq ← fft sworld
⍝ plot freq (⍳size)÷T
```
